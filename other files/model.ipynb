```python
y=d['Yield (Bales/Hectare)'].values
x=d.drop('Yield (Bales/Hectare)',axis=1).values
x
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
l=[]
for i in range(1,25):
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)

  # Convert data to numpy arrays
  X_train = np.array(x_train)
  y_train = np.array(y_train)
  X_test = np.array(x_test)
  y_test = np.array(y_test)

  # Reshape the input data to (samples, timesteps, features)
  n_samples = X_train.shape[0]
  n_timesteps = 1
  n_features = X_train.shape[1]
  X_train = X_train.reshape((n_samples, n_timesteps, n_features))
  X_test = X_test.reshape((X_test.shape[0], n_timesteps, n_features))

  # Define the RNN model
  model = tf.keras.models.Sequential([
      tf.keras.layers.SimpleRNN(64, input_shape=(n_timesteps, n_features)),
      tf.keras.layers.Dense(1)
  ])

  # Compile the model
  model.compile(optimizer='adam', loss='mse')

  # Train the model
  model.fit(X_train, y_train, epochs=100, batch_size=32)

  # Make predictions on the test set
  y_pred = model.predict(X_test)

  # Evaluate the model
  r2 = r2_score(y_test, y_pred)
  print('R2 Score:', r2)
  print("mean squared error:",mean_squared_error(y_test, y_pred))
  l.append((i,r2))
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Convert data to numpy arrays
X_train = np.array(x_train)
y_train = np.array(y_train)
X_test = np.array(x_test)
y_test = np.array(y_test)

# Reshape the input data to (samples, timesteps, features)
n_samples = X_train.shape[0]
n_timesteps = 1
n_features = X_train.shape[1]
X_train = X_train.reshape((n_samples, n_timesteps, n_features))
X_test = X_test.reshape((X_test.shape[0], n_timesteps, n_features))

# Define the RNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.SimpleRNN(64, input_shape=(n_timesteps, n_features)),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, y_train, epochs=200, batch_size=32)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, y_pred)
print('R2 Score:', r2)
print("mean squared error:",mean_squared_error(y_test, y_pred))
```
